### Linear Regression Model Testing

```python
import pandas as pd
import numpy as np
import pickle
import json
```

#### Load Model

```python
with open("Linear_reg_model.pkl", "rb") as f:
    model = pickle.load(f)
model
```

```python
with open("label_encoded_data.json", "r") as f:
    label_enc_data = json.load(f)
label_enc_data
```

```python
column_names = model.feature_names_in_
column_names
```

```python
model.n_features_in_
```

```python
model.coef_
```

### Test Model

```python
age	= 58
gender = "male"
bmi	= 34.89
children = 2
smoker = "yes"
region = "southeast"
```

```python
test_array = np.zeros((1,model.n_features_in_))
test_array
```

```python
test_array[0,0] = age
test_array[0,1] = label_enc_data['gender'][gender]
test_array[0,2] = bmi
test_array[0,3] = children
test_array[0,4] = label_enc_data['smoker'][smoker]

region = f"region_{region}"

region_index = np.where(column_names == region)[0]
test_array[0,region_index] = 1
test_array
```

```python
test_df = pd.DataFrame(test_array, columns = column_names)
test_df
```

```python
predicted_price = model.predict(test_df)[0]
predicted_price
print("Predicted medical insurance price is :",predicted_price)
```
